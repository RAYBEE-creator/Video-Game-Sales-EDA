### Dealing with missing values

```python
# counting the number of missing values in each column
# this clearly shows that the year column has 271 missing values and the publisher column has 58
# since year is a fundamental part of our analysis, we will not drop those missi
vgs.isnull().sum()
```

```python
null_bool = pd.isnull(vgs['Year'])
vgs[null_bool]
```

```python
vgs_1 = vgs.interpolate(method='linear', limit_direction='forward', columns='Year')
vgs_1
```

```python
null_bool = pd.isnull(vgs_1['Year'])
vgs_1[null_bool]
```

```python
vgs_1.iloc[179, :]
```

```python
vgs_1 = vgs_1.astype({'Year': 'int64'})
vgs_1
```

```python
vgs_1.iloc[179, :]
```

```python
# checking for missing values in publishers column
null_bool = pd.isnull(vgs_1['Publisher'])
vgs_1[null_bool]
```

```python
# filling null values
vgs_1 = vgs_1.fillna(method='ffill')
vgs_1
```

```python
# checking if null values have been filled
null_bool = pd.isnull(vgs_1['Publisher'])
vgs_1[null_bool]
```

#### Checking for outliers in the dataset using the describe() function below, which provides a statistical   summary of all the quantitative variables.
#### The years 2017 and 2020 have (3) and (1) counts respectively, are outliers dectected. 
#### The boxplot , histogram and scatter diagrams below gives a clear indication of the outliers.

```python
# Year calue count showing the outliers
vgs_outliers = vgs.Year.value_counts()
print(vgs_outliers)

# summary of year described.
print(vgs['Year'].describe())
plt.figure(figsize=(9, 8));


# pair plot of overall data to compare and detect outliers
sns.pairplot(vgs);

```

#### Box plot

```python
# Box plot to further display outliers
bplt = vgs.boxplot(column ='Year')
```

#### Histogram

```python
vgs['Year'].plot(kind='hist', bins=20, figsize=(12,6), facecolor='purple',edgecolor='black')
```

####  Using the trimming method to deal with the outliers, by first creating an index for all the data points where the year takes 2017 and 2020  values. 
#### Then the index rows are  dropped from the data, while displaying the  summary of the statistics for the variables.

```python
# Indexing the data points
index = vgs[(vgs['Year'] > 2016)].index

# Dropping from the data
vgs_drop =vgs.drop(index, inplace=True)

# Displaying the summary.

vgs['Year'].describe()
```
